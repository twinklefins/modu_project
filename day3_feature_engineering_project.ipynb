```python
# 4. 결측치 확인
df.isnull().sum()
```

```python
# 5. 결측치 처리
# 숫자형 -> 평균으로 채우기
num_cols = df.select_dtypes(include=[np.number]).columns
for col in num_cols:
    df[col].fillna(df[col].mean(), inplace=True)

# 범주형 -> 최빈값으로 채우기
cat_cols = df.select_dtypes(exclude=[np.number]).columns
for col in cat_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)

# 채운 뒤 재확인
df.isnull().sum()
```

```python
# 6. 불필요한 컬럼 제거 - 예측값과 상관없는 ID, 이름, 주소 등 제거
# 예측에 직접적으로 도움 안 되는 컬럼 삭제
drop_cols = ['first', 'last', 'street', 'city', 'state', 'zip', 'dob', 'trans_num']
df.drop(columns=drop_cols, inplace=True, errors='ignore')

# 컬럼명 확인
df.columns
```

```python
# 7. 컬럼명 정리
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace('[^0-9a-zA-Z]+', '_', regex=True)
    .str.replace('_+', '_', regex=True)
    .str.strip('_')
)

# 컬럼명 확인
df.columns
    
```

```python
# 8. 파생변수 만들기 (시간/금액 활용)
# 8-1) 시간 기반
# 날짜 컬럼을 datetime으로 변환
df['trans_dt'] = pd.to_datetime(df['trans_date_trans_time'])

# 시(hour), 요일(dayofweek), 주말 여부
df['hour'] = df['trans_dt'].dt.hour
df['dayofweek'] = df['trans_dt'].dt.dayofweek
df['is_weekend'] = df[['dayofweek']] >= 5

# 시간대 범주
df['time_of_day'] = pd.cut(
    df['hour'],
    bins=[-1, 6, 12, 18, 24],
    labels=['night', 'morning', 'afternoon', 'evening']
)

# 생성된 시간 관련 컬럼 미리보기
df[['trans_date_trans_time', 'hour', 'dayofweek', 'is_weekend', 'time_of_day']].head()
```

```python
# 8-2) 금액 기반
# 금액 사분위수 구하기
q25 = df['amt'].quantile(0.25)
q75 = df['amt'].quantile(0.75)

# 금액 구간화
df['amt_bucket'] = pd.cut(
    df['amt'],
    bins=[-np.inf, q25, q75, np.inf],
    labels=['small', 'mid', 'large']
)

# 생성된 금액 관련 컬럼 미리보기
df[['amt', 'amt_bucket']].head()
```

```python
# 9) 위치/경도 거리 계산 (단순 유클리드 거리)
# 고객 위치(lat, long)와 가맹점 위치(merch_lat, merch_long) 차이 계산
df['location_distance'] = np.sqrt(
    (df['lat'] - df['merch_lat'])**2 +
    (df['long'] - df['merch_long'])**2
)

# 거리 컬럼 확인
df[['lat', 'long', 'merch_lat', 'merch_long']]
```

```python
# 10) 고객별 평균 금액 대비 현재 금액 비율
# 고객별 평균 금액 계산
df['customer_avg_amt'] = df.groupby('cc_num')['amt'].transform('mean')

# 현재 거래 금액 / 평균 금액 비율
df['amt_vs_avg'] = df['amt'] / df['customer_avg_amt']

# 확인
df[['cc_num', 'amt', 'customer_avg_amt', 'amt_vs_avg']].head()
```

```python
# 11) 고객의 하루 거래 횟수
# 날짜만 추출
df['trans_date'] = df['trans_dt'].dt.date

# 고객별 하루 거래 횟수
df['daily_trans_count'] = df.groupby(['cc_num', 'trans_date'])['amt'].transform('count')

# 확인
df[['cc_num', 'trans_date', 'amt', 'daily_trans_count']].head()
```
