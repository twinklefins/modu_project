```python
import random

class Account:
    bank_name = 'SC은행'
    account_count = 0

    def __init__(self, name, balance):
        self.name = name
        self.balance = balance
        self.accountnumber = self.make_account_number()
        self.deposit_count = 0
        self.deposit_list = []
        self.withdraw_list = []
        Account.account_count += 1

    def make_account_number(self):
        first = f"{random.randint(0, 999):03d}"
        second = f"{random.randint(0, 99):02d}"
        third = f"{random.randint(0, 999999):06d}"
        return f"{first}-{second}-{third}"
    
    @classmethod
    def get_account_num(cls):
        print(f"{cls.account_count}개 계좌가 생성되었습니다.")

    def deposit(self, amount):
        if amount >= 1:
            self.balance += amount
            self.deposit_count += 1
            self.deposit_list.append(amount)

            print(f"\n{amount:,}원이 입금되었습니다. 현재 잔액: {self.balance}원")

            if self.deposit_count % 5 == 0:
                interest = int(self.balance * 0.01)
                self.balance += interest
                print(f"\n이자 {interest:,}원이 지급되었습니다. 잔액: {self.balance}원")
        else:
            print(f"1원 이상만 입금할 수 있습니다.")
    
    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            self.withdraw_list.append(amount)
            print(f"\n{amount:,}원이 출금되었습니다. 현재 잔액: {self.balance}원")
        else:
            print("잔액이 부족합니다.")

    def display_info(self):
        print('-' * 40)
        print(f"은행이름: {Account.bank_name}")
        print(f"예금주: {self.name}")
        print(f"계좌번호: {self.accountnumber}")
        print(f"잔고: {self.balance:,}원")

    def deposit_history(self):
        print(f"\n{self.name}님의 입금내역:")
        for amount in self.deposit_list:
            print(f" - {amount:,}원")

    def withdraw_history(self):
        print(f"\n{self.name}님의 출금내역:")
        for amount in self.withdraw_list:
            print(f" - {amount:,}원")

a1 = Account("반짝이", 5000000)
a1.deposit(50000)
a1.deposit_history()

a1.withdraw(30000)
a1.withdraw_history()

a2 = Account("단짝이", 1000000)
a3 = Account("엄마", 2000000)

accounts = [a1, a2, a3]

for acc in accounts:
    if acc.balance >= 1000000:
        acc.display_info()

```
