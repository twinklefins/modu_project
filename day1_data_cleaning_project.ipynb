```python
# 요일 + 시간대별 조합 분석 -> 금요일 밤은 다른가?

grouped = trip.groupby(['weekday', 'hour']).agg({
    'fare_amount': 'mean',
    'trip_distance': 'mean',
    'tip_amount': 'mean'
}).reset_index()

grouped['건수']=trip.groupby(['weekday', 'hour'])['fare_amount'].count().values

grouped.rename(columns={
    'weekday': '요일',
    'hour': '시간',
    'fare_amount': '평균 요금',
    'trip_distance': '평균 거리',
    'tip_amount': '평균팁'
}, inplace=True)

weekday_map = {0:'월', 1:'화', 2:'수', 3:'목', 4:'금', 5:'토', 6:'일'}
grouped['요일'] = grouped['요일'].map(weekday_map)

weekday_order = ['월', '화', '수', '목', '금', '토', '일']
grouped['요일'] = pd.Categorical(grouped['요일'], categories=weekday_order, ordered=True)

grouped.sort_values(by=['요일', '시간'], inplace=True)

grouped.head()
```

```python
pivot = grouped.pivot(index='시간', columns='요일', values='평균 요금')

plt.figure(figsize=(10, 8))
sns.heatmap(pivot, annot=True, fmt=".1f", cmap="YlOrRd")
plt.title("요일-시간대별 평균 요금 Heatmap")
plt.xlabel("요일")
plt.ylabel("시간")
plt.show()
```
